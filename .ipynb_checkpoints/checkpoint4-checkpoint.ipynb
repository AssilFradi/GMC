```python
# deleted because sqft_living (=sqft_above+sqft_basement) is sufficient information.

df=df.drop(columns=["id","date","zipcode","sqft_basement","sqft_lot15","lat","long","sqft_living15"],axis=1)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler
S=StandardScaler()
df1=df.copy()
df1=S.fit_transform(df1)
box_plot=sns.boxplot(data=df1)
box_plot.set(xlabel=["Price","Bathrooms","sqft_living","Grade","sqft_above","sqft_living15"])
```

To begin with the least important features, those would be id, sqft_lot15, long, lat, zipcode, yr_renovated, condition, view, waterfront, and sqft_lot. 

These features have little to no corelation with any of the other features, and thus are fairly useless when wanting to predict other features.

The most important features are price, bedrooms, bathrooms, sqft_living, floors, grade, sqft_above, sqft_basement, yr_built, and sqft_living15.

These features have significant corelations with other features, and are thus usefull when wanting to predict other features.

I have decided to predict the grade of a house, based on the Square-foot living of the house (sqft_living), and on the square footage above the basement (sqft_above).

```python
sns.regplot(x=df["sqft_living"], y=df["price"], fit_reg=False)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
#3. Split your dataset into a training set and a testing set. 

x=df["sqft_living"].values[:,np.newaxis]
y=df["grade"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40) 
```

```python
#4. Apply Linear regression to your training set. 
model=LinearRegression()   
model.fit(x_train,y_train)  
predicted=model.predict(x_test) 


print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
#5. Plot the linear regression. 

plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("Grade")
plt.xlabel("Square-foot Living")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
#5. Measure the performance of linear regression using the testing set. 

print("train Accuracy",model.score(x,y))
print("test Accuracy",model.score(x,y))
```

```python
#6. Apply multi-linear regression and compare it to the linear model. 

x=df[["sqft_above","sqft_living"]]  
y=df["grade"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=40) 

model=LinearRegression() 
model.fit(x_train,y_train) 
predicted=model.predict(x_test) 

print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
print("train Accuracy",model.score(x,y))
print("test Accuracy",model.score(x,y))
```

```python
#7. Apply Polynomial regression and compare it to linear and multilinear regression. 

x= df[["sqft_above", "sqft_living"]]
y= df["grade"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=40)  
lg=LinearRegression()
poly=PolynomialFeatures(degree=3)

x_train_fit = poly.fit_transform(x_train) 
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)

print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

As we compare the results of the three types of regression:


 - the multiple linear regression makes a whole lot of sense considering that the price of a house is rarely dependant on a single factor.

- it becomes evident that polynomial regression here gives the most accurate predictions, followed by multi-linear regression and linear regression.

- The Mean Squared Error declines significantly when we switch from linear regression to multi-linear regression, and to lesser extent as well when we move to polynomial regression.

- When looking at the R squared, we can see that just over half of the output variation can be explained by the model's input with linear regression.

- All these factors show us that the polynomial regression is by far the most accurate, as it produces the smallest MSE, the highest R squared, and the highest train and test accuracy.
